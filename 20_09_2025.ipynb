```python
#compile time/method overloading polymorphism

class Bca_1st_Semester :
    def student(self,a,b=0,c=0,d=0):
        return a+b+c+d

Bca=Bca_1st_Semester()
print(Bca.student(5))
print(Bca.student(10,7,8))
print(Bca.student(-10,4,9,-5))
print(Bca.student(99,0,4,-12))
```

```python
#run time/method overriding
class Animal:
    def sound(self):
        print("Animal makes a sound")

class dog (Animal):
    def sound(self):
        print("bow bow bow")

class cat (Animal):
    def sound(self):
        print("meow meow meow")

class snake (Animal):
    def sound(self):
        print("buuuusssssss")

class cow (Animal):
    def sound(self):
        print("ambbbbaaaa")

obj1=Animal()
obj2=dog()
obj3=cat()
obj4=snake()
obj5=cow()
obj2.sound()
obj1.sound()
obj3.sound()
obj5.sound()
obj4.sound()
```

```python
#operator overloading polymorphism
class book:
    def __init__(self,page):
        self.page=page
    def __add__(self,other):
        return self.page+other.page

obj1=book(200)
obj2=book(300)
print("total pages is ",obj1+obj2)
```

```python
# Abstraction example
from abc import ABC, abstractmethod
class vehicle(ABC):
    def sound(self):
        pass

class car (vehicle):
    def sound(self):
        print("I'm driving a car")

obj=car()
obj.sound()
```

```python
# Abstraction examples
from abc import ABC,abstractmethod
class game(ABC):
    def free_fire(self):
        print("winner winner chicken dinner")

    def BGMI(self):
        pass

    def ludo_king(self):
        print("I'm win the game")

class temple_run(game):
    def BGMI(self):
        print("I got royal pass diamond")


    def chess(self):
        print("I'm a district player")

obj=temple_run()
obj.BGMI()
obj.ludo_king()
obj.chess()
obj.free_fire()
```
